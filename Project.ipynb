```python

years = [2015, 2016, 2017, 2018, 2019]
dfs = {}

for year in years:
    df = pd.read_csv(f"{year}.csv")
    df = df[['Country', 'Happiness Score']]
    df.columns = ['Country', f'Score_{year}']
    dfs[year] = df
```

```python
df_merged = dfs[2015]

for year in years[1:]:
    df_merged = df_merged.merge(dfs[year], on='Country')

print(df_merged.head(10))
```

```python
data = []

for _, row in df_merged.iterrows():
    scores = [row[f'Score_{year}'] for year in years]
    
    # Create sequences
    for i in range(len(scores) - 3):
        X = scores[i:i+3]      # 3 years input
        y = scores[i+3]        # next year
        
        data.append(X + [y])

df_model = pd.DataFrame(data, columns=['t1', 't2', 't3', 'target'])
```

```python
from sklearn.linear_model import LinearRegression

X = df_model[['t1', 't2', 't3']]
y = df_model['target']

model = LinearRegression()
model.fit(X, y)
```

```python
predictions = []

for _, row in df_merged.iterrows():
    X_input = [[
        row['Score_2017'],
        row['Score_2018'],
        row['Score_2019']
    ]]
    
    pred_2020 = model.predict(X_input)[0]
    
    predictions.append({
        'Country': row['Country'],
        'Predicted_2020': pred_2020
    })

df_predictions = pd.DataFrame(predictions)
print(df_predictions.head(10))
```
